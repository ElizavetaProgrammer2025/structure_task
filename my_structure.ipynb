```python
#1.1
# Завдання 1: dict
teams = [
    {"назва": "Леви", "поразки": 2, "перемоги": 5, "нічиї": 1},
    {"назва": "Тигри", "поразки": 1, "перемоги": 6, "нічиї": 2},
    {"назва": "Вовки", "поразки": 3, "перемоги": 4, "нічиї": 3},
]

max_team = max(teams, key=lambda team: team["перемоги"])
print("Команда з найбільшою кількістю перемог (dict):", max_team["назва"])
```

```python
#1.2
class Team:
    def init(self, назва, поразки, перемоги, нічиї):
        self.назва = назва
        self.поразки = поразки
        self.перемоги = перемоги
        self.нічиї = нічиї

teams = [
    Team("Леви", 2, 5, 1),
    Team("Тигри", 1, 6, 2),
    Team("Вовки", 3, 4, 3),
]

max_team = max(teams, key=lambda team: team.перемоги)
print("Команда з найбільшою кількістю перемог (class):", max_team.назва)
```

```python
#2.1
# Завдання 2: dict
students = [
    {"ПІП": "Іваненко Ігор", "курс": 2, "група": "КН-21", "рік_нар": 2003},
    {"ПІП": "Петренко Олена", "курс": 1, "група": "КН-11", "рік_нар": 2004},
    {"ПІП": "Сидоренко Марія", "курс": 3, "група": "КН-31", "рік_нар": 2002},
]

youngest = max(students, key=lambda s: s["рік_нар"])
print("Наймолодший студент (dict):", youngest["ПІП"])
```

```python
#2.2
# Завдання 2: class
class Student:
    def init(self, ПІП, курс, група, рік_нар):
        self.ПІП = ПІП
        self.курс = курс
        self.група = група
        self.рік_нар = рік_нар

students = [
    Student("Іваненко Ігор", 2, "КН-21", 2003),
    Student("Петренко Олена", 1, "КН-11", 2004),
    Student("Сидоренко Марія", 3, "КН-31", 2002),
]

youngest = max(students, key=lambda s: s.рік_нар)
print("Наймолодший студент (class):", youngest.ПІП)
```
